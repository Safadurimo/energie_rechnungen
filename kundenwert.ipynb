```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from pathlib import Path
```

Laden der Wetterdaten

```python
file_name = Path("temperaturdaten","tageswerte_KL_00691_akt","produkt_klima_tag_20230410_20241010_00691.txt")
file_path = os.path.join(os.getcwd(), file_name)
daten = pd.read_csv(file_path, sep=";", usecols=[1, 13])
daten["MESS_DATUM"] = pd.to_datetime(daten["MESS_DATUM"], format="%Y%m%d")
daten.head()
```

Darstellen der Temperaturdaten

```python
daten.columns


daten.plot(x='MESS_DATUM', y=' TMK')
plt.xlabel('Datum')
plt.ylabel('Tagesmitteltemperaturen')
plt.title('Tagesmitteltemperatur in Bremen')
plt.show()
```
